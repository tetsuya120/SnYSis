```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
t=np.arange(-15,15,0.04)
```

```python
def u(t): #escalon unitario
    return (t>=0)*1.0
```

```python
def Π(t,τ):          #puerta
    return u(t+τ/2)-u(t-τ/2)
```

```python
def r(t):
    return t*u(t)
```

```python
def ru(t):
    z=2*Π(t+4.5,3)+2/3*r(t+9)*Π(t+1.5,3)+2/3*r(-t+9)*Π(t-1.5,3)+2*Π(t-4.5,3)
    return z
```

```python
x=Π(t+1.5,1)+2*Π(t,2)+Π(t-1.5,1)
plt.plot(t,x)
plt.grid(True)
```

```python
y=Π(t,4)+Π(t,2)
plt.plot(t,y)
plt.grid(True)
```

```python
z=2*Π(t+4.5,3)+2/3*r(t+9)*Π(t+1.5,3)+2/3*r(-t+9)*Π(t-1.5,3)+2*Π(t-4.5,3)
plt.plot(t,z)
plt.grid(True)
```

```python
plt.plot(t,ru(t))
plt.grid(True)
```
